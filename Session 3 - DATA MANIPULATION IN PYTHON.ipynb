## Chaining Operations

### 1. Filter for respodents over 40 years old
### 2. Select only the columns for degree, income, and sex
### 3. Create a new column 'income_category' that labels income as 'High' if over $50000, otherwise 'Low'
### 4. Sort the results by income in descending orde (highest incomes first)

```python
gss_processed = (
    gss.query("age >40")
        .loc[:,["degree", "income", "sex"]]
        .assign(income_category=lambda x: np.where(x["income"] > 50000, "High", "Low"))
        .sort_values(by="income", ascending=False))
gss_processed
```

## **REMOVING OUTLIERS**

### Inter-Quartile Rule

```python
Q1 = data["Duration"].quantile(0.25)
Q3 = data["Duration"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

data = data[(data["Duration"] >= lower_bound) & (data["Duration"] <= upper_bound)]
print("outlier removed successfully")
```

## **Add the Line of Best Fit to the Scatterplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import linregress

plt.figure(figsize=(7, 5))
sns.regplot(x="Age", y="Credit amount", data=data, scatter_kws={'color':'blue'}, line_kws={'color':'red'})

plt.title("Scatter Plot with Line of Best Fit")
plt.xlabel("Years of Experience")
plt.ylabel("Credit (Kshs)")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import linregress
# Remove missing values
df_clean = data[['Age', 'Credit amount']].dropna()
# Performing regression
slope, intercept, r_value, p_value, std_err = linregress(df_clean['Age'], df_clean['Credit amount'])

plt.figure(figsize=(8, 5))

sns.scatterplot(x=df_clean['Age'], y=df_clean['Credit amount'], color='blue', alpha=0.6, label="Data Points")

x_vals= np.linspace(df_clean['Age'].min(), df_clean['Age'].max(), 100)

y_vals = slope * x_vals + intercept

plt.plot(x_vals, y_vals, color='red', label="Best Fit Line", linewidth=2)

equation_text = f"y = {slope:.2f}x + {intercept:.2f}"

plt.text(df_clean['Age'].min() + 2, df_clean['Credit amount'].max() - 5000, equation_text,
         fontsize=12, color='red', bbox=dict(facecolor='white', alpha=0.5))

plt.xlabel("Age")
plt.ylabel("Credit amount")
plt.title("Scatter Plot of Age vs. Credit amount with Best Fit Line")

plt.legend()

plt.show()
```
