# Mathemetical Operations

# 1. Basic Arithmetic Operations

```python
x = 10
y = 3

addition = x + y  # Addition
subtraction = x - y  #Subtraction
multiplication = x * y #Multiplication
division = x / y       # Division (returns float)
floor_division = x // y #Floor division (integer division)
modulus = x % y        # Modulus (remainder)
exponentiation = x ** y # Exponentiation
```

# 2. Using Built-in Math Functions

```python
import math

square_root = math.sqrt(25)  # Square root
absolute_value = abs(8)      #Absolute value
power = pow(3, 3)            # 3 raised to the power of 3
logarithm = math.log10(100)  # Natural Logarithm (ln)
log_base_10 = math.log10(100) # Logarithm base 10
factorial = math.factorial(5)  # Factorial of 5

print(square_root, absolute_value, power, logarithm, log_base_10, factorial)
```

# 3. **Rounding off**

```python
import math
num = 5.6789
rounded_value = round(num, 3) # Round to 2 decimal places
floor_value = math.floor(num) # Floor (round down)
ceil_value = math.ceil(num)
print(rounded_value, floor_value, ceil_value)
```

# Create DataFrame Using Pandas ( Most common)

```python
import pandas as pd

# Creating a DataFrame
data = {"Name": ["Alice", "Bob", "Charlie", "David", "Dan"],
        "Age": [25, 30, 40, 80,22],
        "County": ["Embu", "Nyeri", "Busia", "Bomet", "Meru"]
       }

df = pd.DataFrame(data)
df
```

```python
df["Salary"] = [50000, 60000, 70000, 80000, 90000]
df['occupation'] = ['Nurse', 'Teacher', 'Politician', 'Doctor', 'Nurse']
df
```

```python
df[df["Age"] > 40]
```

## **Handling Missing Data**
### Introducing a missing value

```python
df.loc[2, "Salary"] = None
df
```

# Suppress Warnings

```python
import warnings

warnings.simplefilter("ignore")
```

```python
print("\nFilling missing values with mean salary:")
df["Salary"].fillna(df["Salary"].mean(), inplace=True)
df
```

# Introduce a missing value (Categorical)

```python
df.loc[2, "occupation"] = None
df
```

## Replace the missing value (Categorical)

```python
df["occupation"].fillna(df["occupation"].mode()[0], inplace=True)
df
```
